**DATA PRE-PROCESSING**

```python
data.drop('education', axis=1, inplace=True)
```

```python
data.head()
```

```python
data.info
```

```python
data.shape
```

```python
drop = ['currentSmoker', 'diaBP']
data.drop(drop, axis=1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
print('Number of outliers to be removed is: ',len(data[data['BMI'] > 43]) + len(data[data['glucose'] > 350]) + len(data[data['totChol'] > 450])+ len(data[data['sysBP'] > 220]))
```

```python
data = data[~(data['totChol'] > 450)]
data = data[~(data['sysBP'] > 220)]
data = data[~(data['BMI'] > 43)]
data = data[~(data['glucose'] > 350)]
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=0,inplace=False)
```
